```python
import transformers
from transformers import pipeline 
```

```python
sentiment_pipeline = pipeline("sentiment-analysis")
```

```python
sentence_1 = "i had a great time at the movie it was really funny"
sentence_2 = "i had a great time at the movie but the parking was terrible"
sentence_3 = "i had a great time at the movie but the parking wasn't great"
sentence_4 = "i went to see a movie"
```

```python
print(sentence_1)
sentiment_pipeline(sentence_1)
```

```python
test=sentiment_pipeline(sentence_1)
[sub["label"] for sub in test]
```

```python
print(sentence_2)
sentiment_pipeline(sentence_2)
```

```python
print(sentence_3)
sentiment_pipeline(sentence_3)
```

```python
print(sentence_4)
sentiment_pipeline(sentence_4)
```

```python
#The default model isn't giving us great results for our neutral sentence.
#However there are loads of models we can choose to use that have been trained on different data with different parameters: https://huggingface.co/models?pipeline_tag=text-classification&sort=downloads&search=sentiment
```

```python
print(sentence_1)
specific_model(sentence_1)
```

```python
print(sentence_2)
specific_model(sentence_2)
```
