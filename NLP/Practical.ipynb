```python
import nltk
from nltk.tokenize import word_tokenize 
from nltk.stem import WordNetLemmatizer,PorterStemmer
from nltk.corpus import stopwords 
import re
import pandas as pd 
```

```python
data = pd.read_csv("tripadvisor_hotel_reviews.csv")
```

```python
data.info()
```

```python
data.head()
```

```python
data['Review'][0]
```

```python
data['review_lowercase']= data["Review"].str.lower() #1st step of preprocessing 
```

```python
data.head()
```

```python
en_stopwords=stopwords.words('english') #2nd step of preprocessing
en_stopwords.remove("not")
```

```python
data["no_stopwords"]=data['review_lowercase'] \
                                .apply(lambda x: ' ' \
                                .join([word for word in x.split() if word \
                                not in (en_stopwords)]))
#The "apply" function in pandas runs a given function on every element in a column
#The "lambda x:" part defines an anonymous function —A quick, unnamed function defined in one line, usually for temporary use.
```

```python
data.head()
```

```python
data['no_stopwords'][0]
```

```python
data['review_no_stopwords_no_punct'] = data \
                                    .apply(lambda x:\
                                    re.sub(r"[*]","star", \
                                    x['no_stopwords'] ), axis=1)
# "axis=1" means we’re applying the function row by row, because each row is treated as a small dictionary of columns
 #replacing * with star                                                 
```

```python
data.head()
```

```python
data['review_no_stopwords_no_punct'] = data. \
                                    apply(lambda x: \
                                    re.sub(r"([^\w\s])","", \
                                    x['review_no_stopwords_no_punct'] ), axis=1)
 #removing punctuations like .,!?;:'"()[] etc
```

```python
data.head()
```

```python
#3rd step of preprocessing tokenization
data['tokenized'] = data.apply(lambda x: \
                               word_tokenize( \
                               x['review_no_stopwords_no_punct'] \
                               ), axis=1)
```

```python
data.head()
```

```python
data['tokenized'][0]
```

```python
ps=PorterStemmer() #4th step of preprocessing is stemming  
```

```python
data["stemmed"]=data["tokenized"]\
                .apply(lambda tokens: \
                [ps.stem(token) for token in tokens])
```

```python
data.head()
```

```python
data['stemmed'][0]

```

```python
#5th step of preprocessing lemmanization
lemmatizer = WordNetLemmatizer()
```

```python
data["lemmatized"] = data["tokenized"] \
                    .apply(lambda tokens: \
                    [lemmatizer.lemmatize(token) \
                     for token in tokens])
```

```python
data.head()
```

```python
data['lemmatized'][0]
```

```python
#6th step of preprocessing is N-grams
tokens_clean = sum(data['lemmatized'], []) #it collects all the list into a single list names token_clean
```

```python
# unigrams: n=1
unigrams = (pd.Series \
            (nltk.ngrams(tokens_clean, 1)) \
            .value_counts()) 
print(unigrams)
```

```python
# bigrams: n=2
bigrams = (pd.Series \
           (nltk.ngrams(tokens_clean, 2)) \
           .value_counts()) 
print(bigrams)
```

```python
ngrams_4 = (pd.Series \
            (nltk.ngrams(tokens_clean, 4)) \
            .value_counts()) 
print(ngrams_4)
```
