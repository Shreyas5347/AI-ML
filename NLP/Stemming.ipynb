```python
from nltk.stem import PorterStemmer #stemming algorithm in NLTK used to reduce words to their root (base) form by removing suffixes
```

```python
ps=PorterStemmer()# create stemmer
```

```python
connect_tokens = ['connecting', 'connected', 'connectivity', 'connect', 'connects']
```

```python
for t in connect_tokens:
    print(t,":",ps.stem(t)) #ps.stem(t) applies the PorterStemmer to a word and returns its stem (root form)
```

```python
learn_tokens = ['learned', 'learning', 'learn', 'learns', 'learner', 'learners']
```

```python
for t in learn_tokens:
    print(t, " : ", ps.stem(t))
```

```python
likes_tokens = ['likes', 'better', 'worse']
for t in likes_tokens:
    print(t, " : ", ps.stem(t))
```

```python
#here wors is gramatically incorrect so we use lemmatization to make sure that the word is in meaningful base form and ensures it does not lose it's meaning
```
