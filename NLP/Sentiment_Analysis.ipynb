```python
from textblob import TextBlob
```

```python
sentence_1 = "i had a great time at the movie it was really funny"
sentence_2 = "i had a great time at the movie but the parking was terrible"
sentence_3 = "i had a great time at the movie but the parking wasn't great"
sentence_4 = "i went to see a movie"
```

```python
print(sentence_1)
sentiment_score=TextBlob(sentence_1)
print(sentiment_score.sentiment.polarity)
```

```python
#TextBlob returns a score between -1 and 1, where negative scores indicate negative sentiment, and positive scores indicate positive sentiment.
```

```python
print(sentence_2)
sentiment_score2=TextBlob(sentence_2)
print(sentiment_score2.sentiment.polarity)
```

```python
print(sentence_3)
sentiment_score3=TextBlob(sentence_3)
print(sentiment_score3.sentiment.polarity)
```

```python
print(sentence_4)
sentiment_score4=TextBlob(sentence_4)
print(sentiment_score4.sentiment.polarity)
```

```python
#VADER
#VADER stands for Valence Aware Dictionary and sentiment Reasoner
#it is especially designed for short, informal, social-media-style text.
#VADER considers things such as capital letters, punctuation, intensifiers, and negation.
#TextBlob is a general NLP library with sentiment analysis, whereas VADER is a specialized sentiment analyzer.
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer 
```

```python
vader_sentiment = SentimentIntensityAnalyzer()
```

```python
print(sentence_1)
print(vader_sentiment.polarity_scores(sentence_1))
#here compound is the overall sentiment score 
#The compond score gives us the overall score for the text and is scored the same as TextBlob between -1 and 1.
```

```python
print(sentence_2)
print(vader_sentiment.polarity_scores(sentence_2))
```

```python
print(sentence_3)
print(vader_sentiment.polarity_scores(sentence_3))
```

```python
print(sentence_4)
print(vader_sentiment.polarity_scores(sentence_4))
```
